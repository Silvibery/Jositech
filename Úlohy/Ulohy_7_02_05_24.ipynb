### Úloha 1 Zašifrovaná správa
Silvia je špiónka. Jej muž musel náhle odísť ale predtým jej nechal šifrovanú
správu:
aldudsbdisfsdimoiretmadvtaoirepmaol
Napísal jej k nem perom poznáku. Šifra bude naša klasika a dostaneš ju
o dva dni poštou. O dva dni prišla poštou šifra na rozlúštenie správy:
01010010111001011100100010001110100
Silvia si pametala že nato aby rozluštiial správu musí:
• priradiť písmenko zo správy k pozícií čísla 1 šifry
• zároven platí že 1 musí maž z prava aj z lava 0

```python
sprava = 'aldudsbdisfsdimoiretmadvtaoirepmaol'
sifra = '01010010111001011100100010001110100'

result = ''

for index, value in enumerate(sifra):

    if value == '1' and sifra[index-1] == '0' and sifra[index+1] == '0':
      result += sprava[index]

print(result)
```

### Úloha 2 Bar plot
Preskúmaj nasledujúci kód . Stĺpcový graf zobrazujúci frekvenciu rôznych
nukleotidov (A, T, C, G) na sekvencií DN <br>ASkús nasledujuce funkcie objektu plt<br>
• plt.xlabe <br>l
• plt.ylab <br>el
• plt.ti <br>tle
Ako by si zmenila farbu frafu na červenú?.

```python
import matplotlib.pyplot as plt

# nas dataset tento raz ako dict
data = {'A':20, 'T':15, 'C':30,'G':35}
nucle = list(data.keys())
hodnoty = list(data.values())

 # vytvorime the bar plot - plt.bar(x_value, y_value)
frekvencie = plt.bar(nucle, hodnoty)

frekvencie[0].set_color('r')
frekvencie[1].set_color('r')
frekvencie[2].set_color('r')
frekvencie[3].set_color('r')

plt.xlabel('nucleotides')
plt.ylabel('freq')
plt.title('Nucleotide frequencies')
plt.show()

# barlist=plt.bar([1,2,3,4], [1,2,3,4])
# barlist[0].set_color('r')
# plt.show()
```

```python
#Riešenie chatGPT - zmena farby všetkých bars:

frekvencie = plt.bar(nucle, hodnoty)

# Loop through each bar and set its color
for bar in frekvencie:
    bar.set_color('r')  # Change 'r' to the desired color

plt.xlabel('nucleotides')
plt.ylabel('freq')
plt.title('Nucleotide frequencies')
plt.show()
```
