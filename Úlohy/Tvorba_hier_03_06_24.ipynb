```python
import random
import matplotlib.pyplot as plt

def hod(pocetKociek=1, pocetStran=6, minHodnota=1):
    '''
    vygeneruje jedno náhodné číslo v rozsahu 1-6
    
    '''
    sucetKociek = 0

    for x in range(pocetKociek): #ked chcem pridat pocet kociek potrebujem iterovat cez zadane cislo prostrednictvom for cyklu
        num = random.randint(minHodnota,pocetStran)
        sucetKociek += num
        
    return sucetKociek 

#mohla by byt samostatnua funkcia na histogram sestrannych kociek tu
katani_list = [] #potrebujem list pre histogram

for hadzanie in range(100000):#pocet hodov
    katani_list.append(hod(pocetKociek=3))

plt.hist(katani_list, range=(3,18), bins = 16)
#range=(pocetKociek, pocetKociek*pocetStran),bins=pocetKociek*pocetStran-(pocetKociek-1)
plt.show
```
